```python
match_new_df = match_df[['id','venue','toss_winner','toss_decision','winner','result','total_runs']]
delivery_df = match_new_df.merge(delivery,on='id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] ==2]
```

```python
delivery_df = delivery_df.groupby('id').apply(lambda x: x.sort_values(['over', 'ball']))

```

```python
delivery_df.reset_index(drop=True, inplace=True)
```

```python
delivery_df['current_score']=delivery_df.groupby('id').cumsum()['total_runs_y']
```

```python
delivery_df['runs_to_win'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['runs_to_win'] = delivery_df['runs_to_win'] +1
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 120 - delivery_df['over']*6 - delivery['ball']
sorted_balls_left = delivery_df.groupby('id')['balls_left'].apply(lambda x: x.sort_values(ascending=False)).reset_index(drop=True)

# Assign the sorted values back to the original DataFrame
delivery_df['balls_left'] = sorted_balls_left



```

```python
delivery_df
```

```python
wickets = delivery_df.groupby('id').cumsum()['is_wicket']
delivery_df['wickets_left'] = 10 - wickets
```

```python
delivery_df['balls_left']=delivery_df['balls_left'].astype('int')
```

**ADDING THE CURR AND RRR**


```python
delivery_df.columns
```

```python
delivery_df['curr'] = (delivery_df['current_score']*6/(120 -delivery_df['balls_left']))
```

```python
delivery_df['rrr'] = ((delivery_df['runs_to_win']*6)/(delivery_df['balls_left']))
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0 
delivery_df['final_result'] = delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team','venue','toss_winner', 'toss_decision','runs_to_win','balls_left','wickets_left','total_runs_x','curr', 'rrr','final_result']]
```

```python
final_df.sample(final_df.shape[0])
```

```python
final_df = final_df[final_df['balls_left'] != 0] 
final_df.dropna(inplace= True)
```

**Encoding data**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
final_df.columns
```

```python



trf1 = ColumnTransformer(
    [
        ('trf1', 'passthrough', [ 'runs_to_win', 'balls_left', 'wickets_left', 'total_runs_x', 'curr',
       'rrr']),  # Columns not to be encoded
        ('trf2', OneHotEncoder(sparse=False, drop='first'), ['batting_team', 'bowling_team', 'venue', 'toss_winner','toss_decision'])
    ]
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps= [
    ('step1',trf1),
    ('step2',LogisticRegression(solver = 'liblinear'))
]
               )


```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 6)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
